```python
#400nm backgrounds

import numpy as np
import matplotlib.pyplot as plt

darkfile = open("/global/homes/r/rjsmith/OpticalPlotting/Vacuum measurements 2-20-19/Background/Dark/2019_02_20__09_10_03.txt")
beamfile = open("/global/homes/r/rjsmith/OpticalPlotting/Vacuum measurements 2-20-19/Background/400nm/2019_02_20__10_33_37.txt")
mirrorfile = open("/global/homes/r/rjsmith/OpticalPlotting/Vacuum measurements 2-20-19/Background/400nm/2019_02_20__10_52_22.txt")

darkdata = np.loadtxt("/global/homes/r/rjsmith/OpticalPlotting/Vacuum measurements 2-20-19/Background/Dark/2019_02_20__09_10_03.txt", skiprows=12)
beamdata = np.loadtxt("/global/homes/r/rjsmith/OpticalPlotting/Vacuum measurements 2-20-19/Background/400nm/2019_02_20__10_33_37.txt", skiprows=12)
mirrordata = np.loadtxt("/global/homes/r/rjsmith/OpticalPlotting/Vacuum measurements 2-20-19/Background/400nm/2019_02_20__10_52_22.txt", skiprows = 12)

darkintensities = np.array([datum[1] for datum in darkdata])
darkangles = np.array([datum[0] for datum in darkdata])
beamintensities = np.array([datum[1] for datum in beamdata])
beamangles = np.array([datum[0] for datum in beamdata])
mirrorintensities = np.array([datum[1] for datum in mirrordata])
mirrorangles = np.array([datum[0] for datum in mirrordata])

beamonlyintensities = np.array([intensity - np.mean(darkintensities) for intensity in beamintensities])
mirroronlyintensities = np.array([intensity - np.mean(mirrorintensities) for intensity in mirrorintensities])
```

```python
plt.plot(beamangles,beamonlyintensities)
plt.plot(mirrorangles,mirroronlyintensities)
plt.xlim(-61,60)
plt.ylim(-10,200)
plt.xlabel("Viewing angle (degrees)")
plt.ylabel("Rate - Dark Rate (Hz)")
plt.title("400nm background in Vacuum")
plt.legend(["No sample","Mirror at -75 degrees"])
```
